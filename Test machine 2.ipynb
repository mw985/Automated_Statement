```python
import pandas as pd
import csv
from openpyxl import load_workbook



# Define the CSV file path and columns to extract
filePath = input('Please enter the path of the folder where the csv monthly statement are stored: ')



# Define the CSV file path
csv_file_path = filePath  # Replace with the actual CSV file path

# Read the CSV file and store its contents in a list
data = []
with open(csv_file_path, 'r', newline='') as csvfile:
    csvreader = csv.reader(csvfile)
    for row in csvreader:
        data.append(row)

# Check if the file has at least 2 rows (to avoid removing all rows)
if len(data) >= 2:
    # Remove the first and last rows
    data = data[1:-1]

    # Write the modified data back to the CSV file
    with open(csv_file_path, 'w', newline='') as csvfile:
        csvwriter = csv.writer(csvfile)
        csvwriter.writerows(data)

    print("First and last rows removed from the CSV file.")
else:
    print("CSV file has less than 2 rows, so no rows were removed.")


    
    
csv_file_path = filePath  # Replace with the actual CSV file path
columns_to_extract = ["Date","Description"]  # Replace with the actual column names
columns2_to_extract = ["Withdraw(HKD)","Deposit(HKD)","Balance(HKD)"]

# Load the existing Excel workbook
excel_file_path = "Integrated_sheet.xlsx"  # Replace with the actual Excel workbook path
wb = load_workbook(excel_file_path)

# Select the worksheet in the Excel workbook
sheet_name = "May"  # Replace with the actual worksheet name
ws = wb[sheet_name]

# Read the CSV file into a DataFrame
df = pd.read_csv(csv_file_path)

# Iterate through the columns and add them to the Excel worksheet
for col_idx, column_name in enumerate(columns_to_extract, start=1):
    # Extract the column data from the DataFrame
    column_data = df[column_name]

    # Iterate through the rows and write the data to the Excel worksheet
    for row_idx, value in enumerate(column_data, start=2):  # Start from row 2 to skip the header
        cell = ws.cell(row=row_idx, column=col_idx)
        cell.value = value
        
for col_idx, column_name in enumerate(columns2_to_extract, start=5):
    # Extract the column data from the DataFrame
    column_data = df[column_name]

    # Iterate through the rows and write the data to the Excel worksheet
    for row_idx, value in enumerate(column_data, start=2):  # Start from row 2 to skip the header
        cell = ws.cell(row=row_idx, column=col_idx)
        cell.value = value

# Save the modified Excel workbook
wb.save(excel_file_path)

print(f"Columns {', '.join(columns_to_extract)} extracted from '{csv_file_path}' and added to '{sheet_name}' in '{excel_file_path}'.")
```
